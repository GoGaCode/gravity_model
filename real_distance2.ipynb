```python
gravity['distance_year_2000'] = gravity['dist'] * gravity['year_2000']
gravity['distance_year_2001'] = gravity['dist'] * gravity['year_2001']
gravity['distance_year_2002'] = gravity['dist'] * gravity['year_2002']
gravity['distance_year_2003'] = gravity['dist'] * gravity['year_2003']
gravity['distance_year_2004'] = gravity['dist'] * gravity['year_2004']
gravity['distance_year_2005'] = gravity['dist'] * gravity['year_2005']
gravity['distance_year_2006'] = gravity['dist'] * gravity['year_2006']
gravity['distance_year_2007'] = gravity['dist'] * gravity['year_2007']
gravity['distance_year_2008'] = gravity['dist'] * gravity['year_2008']
gravity['distance_year_2009'] = gravity['dist'] * gravity['year_2009']
gravity['distance_year_2010'] = gravity['dist'] * gravity['year_2010']
gravity['distance_year_2011'] = gravity['dist'] * gravity['year_2011']
gravity['distance_year_2012'] = gravity['dist'] * gravity['year_2012']
gravity['distance_year_2013'] = gravity['dist'] * gravity['year_2013']
gravity['distance_year_2014'] = gravity['dist'] * gravity['year_2014']
gravity['distance_year_2015'] = gravity['dist'] * gravity['year_2015']
gravity['distance_year_2016'] = gravity['dist'] * gravity['year_2016']
gravity['distance_year_2017'] = gravity['dist'] * gravity['year_2017']
gravity['distance_year_2018'] = gravity['dist'] * gravity['year_2018']
gravity['distance_year_2019'] = gravity['dist'] * gravity['year_2019']

```

```python
gravity['pdi_difference'] = np.abs(gravity['pdi_x'] - gravity['pdi_y'])
gravity['idv_difference'] = np.abs(gravity['idv_x'] - gravity['idv_y'])
gravity['mas_difference'] = np.abs(gravity['mas_x'] - gravity['mas_y'])
gravity['uai_difference'] = np.abs(gravity['uai_x'] - gravity['uai_y'])
gravity['ltowvs_difference'] = np.abs(gravity['ltowvs_x'] - gravity['ltowvs_y'])
gravity['ivr_difference'] = np.abs(gravity['ivr_x'] - gravity['ivr_y'])
```

```python
gravity['pdi_difference'] = gravity['pdi_difference'].fillna(0)
gravity['idv_difference'] = gravity['idv_difference'].fillna(0)
gravity['mas_difference'] = gravity['mas_difference'].fillna(0)
gravity['uai_difference'] = gravity['uai_difference'].fillna(0)
gravity['ltowvs_difference'] = gravity['ltowvs_difference'].fillna(0)
gravity['ivr_difference'] = gravity['ivr_difference'].fillna(0)
```

### calculate the cultural distance as euclidean distance

```python
gravity['cultural_distance'] = np.sqrt(np.square(gravity['pdi_difference']) + np.square(gravity['idv_difference']) + np.square(gravity['mas_difference']) + np.square(gravity['uai_difference']) + np.square(gravity['ltowvs_difference']) + np.square(gravity['ivr_difference']))



```

```python
gravity['cultural_distance']  = 100-gravity['cultural_distance']
```
